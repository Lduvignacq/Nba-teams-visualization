```python
from nba_api.stats.endpoints import shotchartdetail

import pandas as pd
def save_shots_data(YEAR):
    team_id = 0  # change to desired team id
    season_types = ['Regular Season', 'Playoffs']
    frames = []

    for st in season_types:
        df = shotchartdetail.ShotChartDetail(
            team_id=team_id,
            player_id=0,
            context_measure_simple='FGA',
            season_nullable=YEAR,
            season_type_all_star=[st]
        ).get_data_frames()[0]
        df['SEASON_TYPE'] = st
        frames.append(df)

    team_dataframe = pd.concat(frames, ignore_index=True)
    print(team_dataframe.head())
    team_dataframe.to_csv(f'data\\nba_api\\shot_per_season\\shot_chart_team_{YEAR}.csv', index=False)
```

```python
for year in range(2014, 2015):
    YEAR = f"{year}-{year-2000+1:02d}"
    print(f"Processing season: {YEAR}")
    save_shots_data(YEAR)
```
